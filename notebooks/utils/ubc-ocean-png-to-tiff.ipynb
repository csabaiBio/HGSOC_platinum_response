# Select png training images to be converted.

```python
ROOT_DIR = '/tank/WSI_data/Ovarian_WSIs/UBC-OCEAN/'
TRAIN_DIR = '/tank/WSI_data/Ovarian_WSIs/UBC-OCEAN/train_images'

df = pd.read_csv(f"{ROOT_DIR}/train.csv")
# print(df.head())

# make a list of images to be converted according to your criteria

# here I only chose 2 specific ones.
# png_list = [f"{i}.png" for i in [37190, 54506]]

# here I chose all the TMA images
# tma_df = df.loc[df['is_tma']]
# print (tma_df.head())

# create a list of image_id from the filtered rows
image_id_list = df['image_id'].tolist()
# print(image_id_list)

png_list = [f"{i}.png" for i in image_id_list]

print(png_list)
print("Number of images: ",len(png_list))
```

# Convert the specified png training images into a pyramidal image file format such as TIF / TIFF using pyvips
Change the quality of the tiles by changing the 'Q=' value. Values below 20 look pixelated.
File(s) will be in the output directory where they can be downloaded, and properly annotated to make appropriate masks to improve the input data quality.

```python
import pyvips

# loop through the png_list
for png_file in png_list:
    # read a png file from the directory
    image = pyvips.Image.new_from_file(f"{TRAIN_DIR}/{png_file}")
    
    # write to a tiff file with some options
    image.tiffsave(f"/tank/WSI_data/Ovarian_WSIs/UBC-OCEAN/tifs/{png_file[:-4]}.tif", compression="jpeg", Q=95, tile=True, pyramid=True)
```
