```python
# load splits and creaate proc lists iwth slide id etc etc 
import pandas as pd
import os
import glob
```

```python
df_main = pd.read_csv("/mnt/ncshare/ozkilim/BRCA/HGSOC_platinum_responce/HGSOC_TCGA_main.csv")
df_main.head()
# df_main = df_main[["slide_id","label"]]
```

# Loop over all splits folders and create process lists for heatmaps generation script

```python
directory = "/mnt/ncshare/ozkilim/BRCA/HGSOC_platinum_responce/splits/"
for folder_name in os.listdir(directory):
    folder_path = os.path.join(directory, folder_name)
    # Check if it's actually a folder
    if os.path.isdir(folder_path):
        # Search for CSV files ending with '_0.csv' in this folder
        csv_files = glob.glob(os.path.join(folder_path, '*_0.csv'))

        # Load and print each found CSV file
        for file_path in csv_files:
            print(f"Loading file from folder: {folder_name}")
            split = pd.read_csv(file_path)
            split = split.drop('Unnamed: 0', axis=1)
            df_melted = split.melt(var_name='type', value_name='slide_id')
            df_melted = df_melted.dropna(subset=['slide_id'])
            # now join labels 
            df_merged = pd.merge(df_melted, df_main, on='slide_id', how='left')
            df_merged.to_csv("/mnt/ncshare/ozkilim/BRCA/HGSOC_platinum_responce/process_lists/"+folder_name+"_process_list.csv")
```
