```python
wsi_file_path = "/tank/WSI_data/semicol/ds_w_1_merged/slides/UKK_wsi_NoTu_37_region_04.ome.tiff"
```

```python
import openslide
from matplotlib import pyplot as plt

# Path to your WSI file

# Open the WSI file
slide = openslide.OpenSlide(wsi_file_path)

# Read a region of the slide
# These values are an example, you might need to adjust them
# (x, y) is the top left pixel of the region in level 0 reference frame
# (width, height) is the region size
# level is the pyramid level (0 is the highest resolution)
x, y = 0, 0
width, height = 1000, 1000
level = 0
region = slide.read_region((x, y), level, (width, height))

# Convert to an RGB image (removing the alpha channel)
region = region.convert('RGB')

# Display the image
plt.imshow(region)
plt.title('WSI Region')
plt.axis('off')  # Remove axis ticks and labels
plt.show()

# Close the slide after processing
slide.close()
```

```python
import tifffile as tiff
from matplotlib import pyplot as plt

# Path to your OME-TIFF file

# Read the OME-TIFF file
with tiff.TiffFile(wsi_file_path) as tif:
    # Assuming you want to read the first image in the file
    image = tif.pages[0].asarray()

# Display the image
plt.imshow(image, cmap='gray')  # Adjust the colormap as needed
plt.title('OME-TIFF Image')
plt.axis('off')  # Remove axis ticks and labels
plt.show()
```
