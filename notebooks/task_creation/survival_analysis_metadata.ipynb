```python
selected_df["DSS"].value_counts()
selected_df = selected_df.dropna(subset=["DSS"])
```

```python
paths = list(Path('/tank/WSI_data/Ovarian_WSIs/TCGA-OV').glob('*.svs'))
paths[:5]

slide_info = np.array([['-'.join(path.name.split('-')[:3]), path.name.split('.')[0][-3:]] for path in paths])
slide_names, slide_types = slide_info[:, 0], slide_info[:, 1]
```

```python
paths #get path name from these ....
slide_names = np.array(['-'.join(path.name.split('-')[:10])[:-4] for path in paths])
```

```python
patient_DSS_df = selected_df[["bcr_patient_barcode","DSS"]]

# join slide IDs here ...
expanded_data = []

# Iterate through the DataFrame
for _, row in patient_DSS_df.iterrows():
    patient_barcode = row['bcr_patient_barcode']
    DSS = row["DSS"]
    matching_slide_ids = [slide_id for slide_id in slide_names if patient_barcode in slide_id]
    for slide_id in matching_slide_ids:
        expanded_data.append({'case_id': patient_barcode, 'slide_id': slide_id,'DSS':DSS})

# Create the expanded DataFrame
expanded_df = pd.DataFrame(expanded_data)
# Reset the index of the new DataFrame
expanded_df.reset_index(drop=True, inplace=True)

expanded_df["DSS"].isna().any()

```

```python
expanded_df.to_csv("../data/raw_subsets/DSS_platinum_survival.csv",index=None)
```
